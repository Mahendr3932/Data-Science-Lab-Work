# STEP 1B: Load the Three Files into DataFrames

```python
# Load CSV file
df_csv = pd.read_csv("Facebook Friends.csv")

# Load TXT file (tab-separated with encoding fix)
df_txt = pd.read_csv("Facebook Friends.txt", sep="\t", encoding="latin1")

# Load Excel file
df_xlsx = pd.read_excel("Facebook Friends.xlsx")
```

```python
print("CSV Shape:", df_csv.shape)
print("TXT Shape:", df_txt.shape)
print("Excel Shape:", df_xlsx.shape)

print("\nCSV Columns:", df_csv.columns.tolist())
print("\nTXT Columns:", df_txt.columns.tolist())
print("\nExcel Columns:", df_xlsx.columns.tolist())
```

## Data Access Explanation:

The Facebook Friends dataset was successfully loaded from three different file formats: CSV, TXT, and Excel. After loading the data into separate pandas DataFrames, it was observed that all three DataFrames have the same number of rows, columns, and identical column names.

This indicates that the source file format does not affect the actual content of the data, provided the files are read using the correct parameters. The difference lies only in the storage format, not in the data itself. Therefore, any of the file formats can be used for analysis without impacting results.

## Descriptive Statistics Analysis:

Descriptive statistics were generated using the describe() function to summarize the numerical variables in the dataset. This output provides key statistical measures such as count, mean, standard deviation, minimum, maximum, and quartile values for each variable.

The count value indicates the number of valid observations available for analysis. The mean represents the average value, while the standard deviation shows how much the values vary from the mean. The minimum and maximum values indicate the range of the data.

Quartiles (25%, 50%, and 75%) divide the data into four equal parts and help identify the distribution and spread of values. The median (50%) provides a robust measure of central tendency that is less affected by extreme values.

Overall, the describe() output offers a concise summary of the dataset and helps identify patterns, variability, and potential outliers, making it a crucial step in exploratory data analysis.

# STEP 4A: Import Visualization Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# STEP 4B: Histogram of a Numerical Variable (Example: Age)

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

# STEP 4C: Histogram of Another Numerical Variable (Friends Count)

```python
plt.figure(figsize=(8,5))
plt.hist(df['Friends'], bins=15)
plt.xlabel('Number of Friends')
plt.ylabel('Frequency')
plt.title('Histogram of Number of Friends')
plt.show()
```

```python
# STEP 4D: Frequency Plot (Count Plot – Example: Gender)
```

```python
df.columns = df.columns.str.strip()
# Column names were cleaned to remove extra spaces before visualization.
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Frequency Plot of Gender')
plt.show()
```
